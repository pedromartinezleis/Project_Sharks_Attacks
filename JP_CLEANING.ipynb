```python
#Esta es la función. Recibe el valor de las filas de la columna Type. Devuelve "Yes" o "No" dependiendo de lo que había originalmente.
 '''
def f(row):
    if row['Type'] == "Boating":
        val = "Yes"
    elif row['Type'] == "Unprovoked":
        val = "No"
    elif row['Type'] == "Invalid":
        val = "No"
    elif row['Type'] == "Provoked":
        val = "Yes"
    elif row['Type'] == "Questionable":
        val = "Yes"
    elif row['Type'] == "Sea Disaster":
        val = "Yes"
    elif row['Type'] == "Boat":
        val = "Yes"
    elif row['Type'] == "Boatomg":
        val = "Yes"
    elif row['Type'] == "Unknown":
        val = "Unknown"
    return val
'''
```

```python
#y veo si funciona
df['Provoked'] = df.apply(fc.f, axis=1)
```

```python
#Ahora quiero que las muertes sean simplemente "N" o "Y". Transformo los valores usando la misma función. "NaN", "M", "Unknown" y "2017" pasarás a ser "Unknown". EL resto serán Y o N según proceda.
```

```python
def y(row):
    if row['Fatal (Y/N)'] == "N":
        i = "N"
    elif row['Fatal (Y/N)'] == "Y":
        i = "Y"
    elif row['Fatal (Y/N)'] == "nan":
        i = "Unknown"
    elif row['Fatal (Y/N)'] == "M":
        i = "Unknown"
    elif row['Fatal (Y/N)'] == "Unknown":
        i = "Unknown"
    elif row['Fatal (Y/N)'] == "2017":
        i = "Unknown"
    elif row['Fatal (Y/N)'] == " N":
        i = "N"
    elif row['Fatal (Y/N)'] == "N ":
        i = "N"
    elif row['Fatal (Y/N)'] == ", ":
        i = "Unknown"
    else:
        i = "Y"
    return i
```

```python
df['Muerte'] = df.apply(fc.y, axis=1)
```
